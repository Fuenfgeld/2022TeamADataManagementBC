```python
# Datensatz laden
import sklearn.datasets as ds
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

diabetesDS = ds.load_breast_cancer(as_frame=True)
```

```python
# Allgemeine Inhalte ausgeben
for i in diabetesDS:
    print(i)
```

```python
# Beschreibende Daten aufrufen -- Metadaten
print(diabetesDS.DESCR)
```

```python
label_names = diabetesDS['target_names'] 
labels = diabetesDS['target'] 
feature_names = diabetesDS['feature_names'] 
features = diabetesDS['data'] 

print("Label Names:\n  %s .\n\n" % (label_names)) 
print("labels:\n  %s .\n\n" % (labels)) 
print("Feature Names:\n  %s .\n\n" % (feature_names)) 
print("Features:\n  %s .\n\n" % (features)) 
```

```python
# Daten und Zieldaten(Brustkrebs Ja/Nein) zu Dataframe zusammenfassen und die ersten 5 Zeilen ausgeben
df = pd.DataFrame(diabetesDS.data, columns=diabetesDS.feature_names)
df['target'] = diabetesDS.target

df.head()
```

```python
# Größe des Datensatzes 
df.shape
```

```python
# Weitere Information über Datentypen der Spalten auslesen
df.info()
```

```python
# Anzahl, Mittelwert, Standardabweichung, Minimum und Maximum für numerische Variablen zusammenfassen.
df.describe()
```

```python
# Anzahl der eindeutigen Werte für jede Variable 
df.nunique(axis=0)
```

```python
# Ersetzen von Leerwerten (mit Leerzeichen) durch NAN
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# Gibt es Dopplungen?
print("Number of Duplicated Rows", df.duplicated(df.columns).sum())
```

```python
df.isnull().sum()
```

```python
# Korrelationsmatrix
corr = df.corr()

plt.figure(figsize=(15, 13))
sns.heatmap(corr, annot=False)
```

```python
# Histogramm
df['mean radius'].plot(kind='hist', bins=20, figsize=(12,6), facecolor='grey',edgecolor='black')
```

```python
# Verteilung Maligne(0) und Benigne(1)
sns.countplot(x="variable", hue= "value", data=pd.melt(df[["target"]]))
```

```python
# Zusammenhang für Maligne(0) oder Benigne(1) mit dem gemessenen mittleren Radius als Scatterplot dargestellt
df.plot(kind='scatter', y='mean radius', x='target')

#df.plot(kind='scatter', y='mean radius', x='mean area')
```

```python
# Alle Kategorien welche die Mittelwerte abbilden, in ihrer Abhängigkeit zueinander vergleichen, zusätzliches farblich kodiertes Mapping auf Target

meansDF = df[['mean radius',
 'mean texture',
 'mean perimeter',
 'mean area',
 'mean smoothness',
 'mean compactness',
 'mean concavity',
 'mean concave points',
 'mean symmetry',
 'mean fractal dimension',
 'target']]

sns.set_theme(style="ticks")

sns.pairplot(meansDF, hue="target")
```
